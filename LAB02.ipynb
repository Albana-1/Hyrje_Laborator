# **2. Baza teorike**


Densiteti jepet nga formula 

 $d = \frac{m}{V}$.

 ku:
 * **d** - densiteti
 * **m** - masa e trupit
 * **V** - vellimi i trupit

**Teoria**

Densiteti është një madhësi fizike që tregon raportin midis masës dhe vëllimit të një trupi. Ai është karakteristik për çdo material dhe nuk varet nga forma apo madhësia e trupit.
